# Scope of Variable : 
In Python, the scope of a variable refers to the region of the program where a variable is recognized and can be accessed or modified.

It defines how fast a variable can be accessed and modified.

- Python follow LEGB rules

**Types of Variable Scopes in Python (LEGB Rule)**

| Scope Type | Full Form     | Description                                                                         | Example                          |
| ---------- | ------------- | ----------------------------------------------------------------------------------- | -------------------------------- |
| **L**      | **Local**     | Variables declared **inside a function**. Accessible **only within** that function. | Inside a function                |
| **E**      | **Enclosing** | Variables in the **outer function** for **nested functions**.                       | Function inside another function |
| **G**      | **Global**    | Declared **outside all functions**, accessible throughout the module.               | Top-level variables              |
| **B**      | **Built-in**  | Predefined names in Python (e.g., `len`, `sum`, etc.).                              | Pythonâ€™s built-in scope          |

**Global Scope Variable**

- If we declare any variable in main space then it is known as global scope of variable
- We can access global scope varible at main space as well as functional space.
- We can not modify global scope variable inside function space/method area , in order to do that we have to use `global` keyword.

```python

a= 10 # Global Space 
def main():
    print("Inside functional space  :",a) # global space accessed inside function
print("At main space :",a) ## Global Space outside function
main()
```

# Local Scope : 
- If we declare a var inside a function then it is known as a local scope of variable.
- we can not access local scope of var at main space, it can be accessed inside of the function only

```python
def inner():
        x = "local"  # Local Scope
        print("Inner:", x)
inner()
#print("Outer:", x)  # This will raise an error because x is not defined in the global scope
```

# Enclosed / Non local Scope of Variable

- If we declare any var inside nested function then it is consider as Enclosed/Non local Scope of variable.
- We can access this eclosed var at non local and local.


```python
#ðŸ”¸ Example Demonstration (LEGB Rule)

x = "global"  # Global Scope

def outer():
    x = "enclosing"  # Enclosing Scope

    def inner():
        x = "local"  # Local Scope
        print("Inner(local space):", x)

    inner()
    print("Outer (non local space): ", x)

outer()
print("Global:", x)
```

# Non local

- If we try to modify the non local variable inside the local space by default it is not possible ,in order to do that we have to use `non local` keyword.
