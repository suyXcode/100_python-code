```python
class school:
    sname="DPS"
    phno=1234567890
    address="Delhi"
    def __init__(self,name,age,rollno,sid):
        self.name=name
        self.age=age
        self.rollno=rollno
        self.sid=sid 
    def display(self):
        print(f"School Name : {school.sname}")
        print(f"School Phone Num : {school.phno}")
        print(f"School Address : {school.address}")
        print(f"Student Name : {self.name}")
        print(f"Student Age : {self.age}")
        print(f"Student Roll num : {self.rollno}")
        print(f"Student ID : {self.sid}")
class house(school):
    hcap="Suyash Singh"
    prize=100
    members=20
    def __init__(self,name,age,rollno,sid,position):
        super().__init__(name,age,rollno,sid)
        # school().__init__(name,age,rollno,sid)
        self.position=position
    def display(self):
        print(f"House Captain : {house.hcap}")
        print(f"Prize : {house.prize}")
        print(f"Members : {house.members}")
        super().display()
        # school().display()
        print(f"Position : {self.position}")
s1=house("Ankit verma",18,1,1,23)
s1.display()
```

```python
# # create a class hospital which have four class members four object members a child class named as cardiology is inherit all properties from parent class
# 1. create two method to display both class members and object members
# create another method to helps to change phno of patient
# create a method which help us to main changes in the hospital location


class Hospital:
    # ------- Class Members (Common for all objects) -------
    hospital_name = "City Care Hospital"
    location = "Lucknow"
    established = 1995
    specialization = "Multi-speciality"

    def __init__(self, patient_name, patient_age, patient_disease, patient_phno):
        self.patient_name = patient_name
        self.patient_age = patient_age
        self.patient_disease = patient_disease
        self.patient_phno = patient_phno


    @classmethod
    def display_class_members(cls):
        print("\n--- Hospital Information ---")
        print("Hospital Name:", cls.hospital_name)
        print("Location:", cls.location)
        print("Established Year:", cls.established)
        print("Specialization:", cls.specialization)

    def display_object_members(self):
        print("\n--- Patient Information ---")
        print("Name:", self.patient_name)
        print("Age:", self.patient_age)
        print("Disease:", self.patient_disease)
        print("Phone No:", self.patient_phno)


    def update_phone(self, new_phno):
        self.patient_phno = new_phno
        print(f"Phone number updated to {new_phno} for {self.patient_name}")

    
    @classmethod
    def update_location(cls, new_location):
        cls.location = new_location
        print(f"Hospital location updated to {new_location}")


# ------- Child Class -------
class Cardiology(Hospital):
    def __init__(self, patient_name, patient_age, patient_disease, patient_phno):
        super().__init__(patient_name, patient_age, patient_disease, patient_phno)


p1 = Cardiology("Amit Sharma", 45, "Heart Disease", "9876543210")
p2 = Cardiology("Riya Verma", 52, "Cardiac Arrest", "9123456780")


Cardiology.display_class_members()


p1.display_object_members()
p2.display_object_members()


p1.update_phone("9999999999")


Cardiology.update_location("Kanpur")


Cardiology.display_class_members()
p1.display_object_members()
```
