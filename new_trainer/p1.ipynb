```python
#Wap to extract all upper 
s = "PyTHon"
count=0
i=0
out=""
while i< len(s):
    if s[i].isupper(): # if  "A"<= s[i]<= "Z":
        print(s[i])
        out += s[i]
        count+=1
    i+=1
print("Count :",count)
print("Output:",out)
```

```python
#wap extract vowel
s = "DaTAScieNCe"
count=0
i=0
out=""
while i< len(s):
    if s[i] in "AEIOUaeiou": 
        print(s[i])
        out += s[i]
        count+=1
    i+=1
print("Count :",count)
print("Output:",out+str(count))
```
